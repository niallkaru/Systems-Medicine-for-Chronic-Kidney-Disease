```python
#### FITTING ####
X0 = np.array([0, 6, 17])
params_fit = spm.params_fsm(lam1=0.9,lam2=0.8,mu1=0.3,mu2=0.3,K=1e6,k1=1e9,\
                            k2=1e9,beta1=470*min2day,beta2=70*min2day,\
                            beta3=240*min2day,alpha1=940*min2day,alpha2=510*min2day,\
                            gamma=2,n=0.0003,h=50,r=100,q=1e5)
state_fit = spm.state_fsm(6,0,0,0,17)
initial_guess = [0.003,50,100,1e3]
cell_fit = fsm.fibrosis_senescence_model(params_fit,state_fit)

fitting_data = pd.read_excel(r'C:\Users\niall\Documents\MPhys\Systems-Medicine-for-Chronic-Kidney-Disease\cell_numbers_for_model.xlsx',skiprows=0)
t_data = fitting_data.iloc[:, 0].to_numpy()
M_data = fitting_data.iloc[:, 1].to_numpy()
F_data = fitting_data.iloc[:, 3].to_numpy()
S_data = fitting_data.iloc[:, 5].to_numpy()
result = cell_fit.fit_data(t_data, M_data, F_data, S_data, X0, initial_guess,method='trust-constr',bounds = [(0, None), (0, None), (0, None),(100,1000)])

sol_fit = solve_ivp(cell_fit.change_in_m_f_to_int, (t_data[0], t_data[-1]), X0, t_eval=t_data, method='Radau')
M_fit, F_fit, S_fit = sol_fit.y

plt.figure(figsize=(8, 6))

# Plot experimental data as markers
plt.plot(t_data, M_data, 'o', markersize=4, label="Data M",color='red')
plt.plot(t_data, F_data, 'o', markersize=4, label="Data F",color='darkgoldenrod')
plt.plot(t_data, S_data, 'o', markersize=4, label="Data S",color='seagreen')

# Plot fitted model simulation as lines
plt.plot(t_data, M_fit, '-', label="Fitted M",color='lightcoral')
plt.plot(t_data, F_fit, '-', label="Fitted F",color='goldenrod')
plt.plot(t_data, S_fit, '-', label="Fitted S",color='mediumaquamarine')
#plt.yscale('log')
#plt.xscale('log')
plt.xlabel("Time")
plt.ylabel("Cell Population")
plt.title("Fitted Simulation vs. Experimental Data")
plt.legend()
plt.tight_layout()
plt.show()
```
